```python
from radis import calc_spectrum
s = calc_spectrum(1900,           # wmin (minimum wavelength/wavenumber) 400
                  2300,           # wmax (maximum wavelength/wavenumber) 12500
                  molecule='CO',
                  isotope='1,2,3',
                  pressure=0.0001,  # Bar
                  Tgas=294.15,      # K
                  Tvib=None,        # vibrational temperature (K)
                  Trot=None,        # rotational temperature (K)
                  mole_fraction=1,
                  path_length=10,   # cm
                  databank='hitran',
                  wstep='auto',
                  use_cached=True,
                  export_lines=False,
                  )
s.apply_slit(0.5, 'nm')
s.plot('transmittance_noslit', show=True)
```

```python
from radis import sPlanck

s = sPlanck(wavenum_min=1900,  # 400
            wavenum_max=2300,  # 12500
            T=1700,            # blackbody temperature (K) - Globar
            eps=1,             # grey-body emissivity
            medium='air')
s.plot()
```
